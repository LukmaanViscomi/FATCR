```python
from pathlib import Path
import json

feedback_path = Path(r"C:\Users\lukmaan\Msc_in_AI\FATCR\factr_verdict_feedback.jsonl")

rows = []
with feedback_path.open(encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        rows.append(json.loads(line))

print("Total feedback entries:", len(rows))
print("First 3 rows:")
for r in rows[:3]:
    print(r)
```

```python
import matplotlib.pyplot as plt

# Data
metrics = ["Coverage (%)", "Mean items/claim", "Strong-match rate (%)"]
islam = [93.4, 2.64, 12.1]
christian = [97.8, 2.82, 22.0]

x = range(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))

# Bars
ax.bar([i - width/2 for i in x], islam, width, label="Islam evidence")
ax.bar([i + width/2 for i in x], christian, width, label="Christian evidence")

# Labels and title
ax.set_ylabel("Value")
ax.set_xticks(list(x))
ax.set_xticklabels(metrics, rotation=15, ha="right")
ax.set_title("Evidence Retrieval Summary over 91 Claims (FACTR)")
ax.legend()

# Annotate bars with values
for i, val in enumerate(islam):
    ax.text(i - width/2, val + (0.5 if val > 10 else 0.1), f"{val:.1f}", ha='center', va='bottom', fontsize=8)

for i, val in enumerate(christian):
    ax.text(i + width/2, val + (0.5 if val > 10 else 0.1), f"{val:.1f}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
fig_path = "/mnt/data/factr_evidence_infographic.png"
plt.savefig(fig_path, dpi=300)
fig_path
```
